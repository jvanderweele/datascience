Begin Production Code:

```python
###Get hate tweets, 100 at a time, up to 10,000
num_needed = 10000
tweet_list_hate = []
last_id = -1 # id of last tweet seen
while len(tweet_list_hate) < num_needed:
    try:
        new_tweets = api.search(q = 'hate', count = 100, max_id = str(last_id - 1))
    except tweepy.TweepError as e:
        print("Error", e)
        break
    else:
        if not new_tweets:
            print("Could not find any more tweets!")
            break
        tweet_list_hate.extend(new_tweets)
        last_id = new_tweets[-1].id
```

```python
###Get love tweets, 100 at a time, up to 10,000
num_needed = 10000
tweet_list_love = []
last_id = -1 # id of last tweet seen
while len(tweet_list_love) < num_needed:
    try:
        new_tweets = api.search(q = 'love', count = 100, max_id = str(last_id - 1))
    except tweepy.TweepError as e:
        print("Error", e)
        break
    else:
        if not new_tweets:
            print("Could not find any more tweets!")
            break
        tweet_list_love.extend(new_tweets)
        last_id = new_tweets[-1].id
```

Here is where we begin flag creation for our criteria:

```python
#######################################
#HATE Tweets Dataset Quantification
#######################################



proexamples = ["hate you", "hate him", "hate her", #Singular 2nd and 3rd person objective pronouns
               ###After reviewing tweets, I'm deciding to keep "hate them" as personal pronoun because many of tweets
               ###I have reviewed clearly intend "them" to be people, and not shoes.
                "hate them", #Plural 2nd and 3rd Person objective pronouns (excluded you as it is in first set)
                 "hate someone", "hate everyone", "hate somebody", "hate everybody" #Singular Indefinite excluding 
               #ones like anybody/nobody
                ]
negexamples = ["hate your", "hate himsel", "hate hersel", "hate themsel", "hate someones", 
               "hate everyones", "hate someone'", "hate everyone'"] #This is a list of problematic statements we 
                #are excluding.

def PronounFlag(input_str):
    """This function will flag hate pronoun examples while excluding ones we want to negate"""
    if (any(x in input_str for x in proexamples) and any(x in input_str for x in negexamples)==False):
        return 1 #True
    if any(x in input_str for x in negexamples):
        return 0
    return 0 #False

unkexamples = ["hate my"] #Unknown if toward person, object, event

def UnknownFlag(input_str):
    """This function will flag tweets that may be unknown if toward person, object, or event"""
    if any(x in input_str for x in unkexamples):
        return 1 #True
    return 0 #False

atsymbol = ["@"] #At symbol represents another twitter account (could be a person)

def AtSymbol(input_str):
    """This function will flag tweets that contain an @ symbol which represents another twitter account"""
    if any(x in input_str for x in atsymbol):
        return 1 #True
    return 0 #False

# ##For sanity check on lower case interpretation:
# def OriginalTweet(input_str):
#     """This function will help check that tweets are lowercase"""
#     return(input_str)

##Run the functions for each row of dataframe. Also we just make the tweet lowercase so not to have to deal with
##capitalization issues.
caseTweet = hateDF.Tweet
##Original function is hateDF.Tweet.apply(flag)
hateDF.loc[:, 'PronounFlag'] = caseTweet.str.lower().apply(PronounFlag) #Locate/Label my new DF column
hateDF.loc[:, 'UnknownFlag'] = caseTweet.str.lower().apply(UnknownFlag) #Locate/Label my new DF column
hateDF.loc[:, 'AtSymbolFlag'] = caseTweet.str.lower().apply(AtSymbol) #Locate/Label my new DF column

#####The below is a sanity check to ensure that tweets are being interpreted as lower case. Additionally, 
#####I included some lowercase/capitalized combinations to check myself and it worked as expected.
#hateDF.loc[:, 'LowerCaseTweet'] = caseTweet.str.lower().apply(OriginalTweet) #Locate/Label my new DF column


#print(hateDF.PronounFlag, hateDF.Tweet) #If we want to see some data
```
